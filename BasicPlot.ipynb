```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 30, 40])

plt.plot(x, y, color = 'r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

```python
days = list(range(0, 22, 3))
temperatureC = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]

plt.plot(days, temperatureC)
plt.show()
```

```python
x = np.arange(1, 11)
y = np.exp(x)

plt.plot(x, y)
plt.show()
```
